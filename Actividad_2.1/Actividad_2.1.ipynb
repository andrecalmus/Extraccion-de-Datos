Filtro a: Los registros de 5 host names diferentes

```python
#Filtro a

filtro_a = df[(df["NOMBRE ANFITRION"] == "Noelle") | (df["NOMBRE ANFITRION"] == "Mike") | (df["NOMBRE ANFITRION"] == "Gaby") | (df["NOMBRE ANFITRION"] == "Oscar") | (df["NOMBRE ANFITRION"] == "Jasmine")]
```

```python
#Convertir archivo filtrado a CSV

filtro_a.to_csv("filtro_a.csv")
```

Filtro b: Los registros de los host que se hallan unido a Airbnb despúes del año 2020

```python
#Conversión de tipo de dato de columna

df["FECHA INICIO ANFITRION"] = df["FECHA INICIO ANFITRION"].astype("datetime64[s]")
```

```python
#Filtro b

filtro_b = df[(df["FECHA INICIO ANFITRION"] >= "2020-01-01")]
```

```python
#Convertir archivo filtrado a CSV

filtro_b.to_csv("filtro_b.csv")
```

Filtro c: Los registros de los host que responden a más tardar en un día

```python
#Filtro c

filtro_c = df[(df["TIEMPO DE RESPUESTA"] == "within an hour") | (df["TIEMPO DE RESPUESTA"] == "within a few hours") | (df["TIEMPO DE RESPUESTA"] == "within a day")]
```

```python
#Convertir archivo filtrado a CSV

filtro_c.to_csv("filtro_c.csv")
```

Filtro d: Los registros de las filas múltiplos de 200... (400, 600, 800, etc)

```python
#Filtro d

filtro_d = df.iloc[range(0, len(df), 200)]
```

```python
#Convertir archivo filtrado a CSV

filtro_d.to_csv("filtro_d.csv")
```

Filtro e: Los registros de los tipos de cuarto Entire home y Private room

```python
#Filtro e

filtro_e = df[(df["TIPO DE CUARTO"] == "Private room") | (df["TIPO DE CUARTO"] == "Entire home/apt")]
```

```python
#Convertir archivo filtrado a CSV

filtro_e.to_csv("filtro_e.csv")
```

Filtro f: Los registros que a lo mucho cuenten con 4 camas

```python
#Filtro f

filtro_f = df[df["CAMAS"] <= 4]
```

```python
#Convertir archivo filtrado a CSV

filtro_f.to_csv("filtro_f.csv")
```

Filtro g: Los registros que sean superhost

```python
#Filtro g

filtro_g = df[df["SUPERANFITRION"] == "t"]
```

```python
#Convertir archivo filtrado a CSV

filtro_g.to_csv("filtro_g.csv")
```

Filtro h: Los registros de las columnas impares

```python
#Filtro h

filtro_h = df.iloc[:, 1::2]
```

```python
#Convertir archivo filtrado a CSV

filtro_h.to_csv("filtro_h.csv")
```
