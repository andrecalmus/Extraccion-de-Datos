Filtro a: Los registros de los host cuyo host response rate sea mayor a 95% y que no sean superhost

```python
#Eliminar un signo de una columna

df["TASA DE RESPUESTA"]=df["TASA DE RESPUESTA"].str.replace('%', '')
```

```python
#ConversiÃ³n de tipo de dato de columna

df["TASA DE RESPUESTA"]=df["TASA DE RESPUESTA"].astype(float)
```

```python
#Filtro a

filtro_a = df[(df["TASA DE RESPUESTA"] >= 95) & (df["SUPERANFITRION"] == "f")]
```

```python
#Convertir archivo filtrado a CSV

filtro_a.to_csv("filtro_a.csv")
```

Filtro b: Los registros cuyas propiedades no tengan 1 baÃ±o

```python
#Filtro b

filtro_b = df[df["BAÃ‘OS"] != 1]
```

```python
#Convertir archivo filtrado a CSV

filtro_b.to_csv("filtro_b.csv")
```

Filtro c: Los registros que tengan un precio entre $100 y $200

```python
#Eliminar un signo de una columna

df["PRECIO"]=df["PRECIO"].str.replace('$', '')
df["PRECIO"]=df["PRECIO"].str.replace(',', '')
```

```python
#ConversiÃ³n de tipo de dato de columna

df["PRECIO"]=df["PRECIO"].astype(float)
```

```python
#Filtro c

filtro_c = df[(df["PRECIO"] >= 100) & (df["PRECIO"] <= 200)]
```

```python
#Convertir archivo filtrado a CSV

filtro_c.to_csv("filtro_c.csv")
```

Filtro d: Los registros que sean instant bookable y que sean superhost

```python
#Filtro d

filtro_d = df[(df["RESERVA INSTANTANEA"] == "t") & (df["SUPERANFITRION"] == "t")]
```

```python
#Convertir archivo filtrado a CSV

filtro_d.to_csv("filtro_d.csv")
```

Filtro e: Los registros de los tipos de cuarto: Hotel room o Private room

```python
#Filtro e

filtro_e = df[(df["TIPO DE CUARTO"] == "Private room") | (df["TIPO DE CUARTO"] == "Hotel room")]
```

```python
#Convertir archivo filtrado a CSV

filtro_e.to_csv("filtro_e.csv")
```

Filtro f: Los registros que tengan mÃ¡s de 4 accommodates

```python
#Filtro f

filtro_f = df[df["CAPACIDAD"] > 4]
```

```python
#Convertir archivo filtrado a CSV

filtro_f.to_csv("filtro_f.csv")
```
